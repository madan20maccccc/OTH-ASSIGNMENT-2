# **Data Description**

```python
import pandas as pd

# Load feature matrices
X_train = pd.read_csv("X_train.txt", delim_whitespace=True, header=None)
X_test  = pd.read_csv("X_test.txt", delim_whitespace=True, header=None)

# Load labels
y_train = pd.read_csv("y_train.txt", header=None, names=["activity"])
y_test  = pd.read_csv("y_test.txt", header=None, names=["activity"])

# Check shapes
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape :", X_test.shape)
print("y_test shape :", y_test.shape)
```

# **Data Preprocessing**

```python
# Load feature names
features = pd.read_csv(
    "features.txt",
    sep="\s+",
    header=None,
    names=["index", "feature_name"]
)

# Extract only the feature names
feature_names = features["feature_name"].values

# Assign column names
X_train.columns = feature_names
X_test.columns = feature_names

# Verify
print("First 10 feature names:")
print(X_train.columns[:10])

print("\nCheck dataframe head:")
X_train.head()
```

```python
# Attach labels to feature data
train_df = X_train.copy()
train_df["activity"] = y_train["activity"]

test_df = X_test.copy()
test_df["activity"] = y_test["activity"]

# Check final shapes
print("Train dataframe shape:", train_df.shape)
print("Test dataframe shape :", test_df.shape)

# Save clean CSV files
train_df.to_csv("har_train_clean.csv", index=False)
test_df.to_csv("har_test_clean.csv", index=False)

print("\nCSV files created successfully!")
```

```python
import matplotlib.pyplot as plt

# Count activity labels
activity_counts = train_df["activity"].value_counts().sort_index()

# Plot
plt.figure()
plt.bar(activity_counts.index, activity_counts.values)
plt.xlabel("Activity Label")
plt.ylabel("Number of Samples")
plt.title("Activity Distribution in Training Data")
plt.show()

print(activity_counts)
```

```python
# Load activity labels
activity_labels = pd.read_csv(
    "activity_labels.txt",
    sep="\s+",
    header=None,
    names=["label", "activity_name"]
)

# Create mapping dictionary
activity_map = dict(zip(activity_labels.label, activity_labels.activity_name))

# Add activity name column
train_df["activity_name"] = train_df["activity"].map(activity_map)

# Verify
train_df[["activity", "activity_name"]].head()
```

# **Expolatory Data Analysis - EDA**

```python
# Count activities by name
activity_name_counts = train_df["activity_name"].value_counts()

# Plot
plt.figure()
plt.bar(activity_name_counts.index, activity_name_counts.values)
plt.xlabel("Activity")
plt.ylabel("Number of Samples")
plt.title("Activity Distribution in Training Data")
plt.xticks(rotation=45)
plt.show()

print(activity_name_counts)
```

```python
from sklearn.decomposition import PCA

# Separate features and labels
X_pca = train_df.drop(columns=["activity", "activity_name"])
y_pca = train_df["activity_name"]

# Apply PCA
pca = PCA(n_components=2)
X_pca_2d = pca.fit_transform(X_pca)

# Create PCA dataframe
pca_df = pd.DataFrame(X_pca_2d, columns=["PC1", "PC2"])
pca_df["activity"] = y_pca.values

# Plot
plt.figure()
for activity in pca_df["activity"].unique():
    subset = pca_df[pca_df["activity"] == activity]
    plt.scatter(subset["PC1"], subset["PC2"], label=activity, alpha=0.5)

plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA Visualization of HAR Dataset (Training Data)")
plt.legend()
plt.show()

# Explained variance
print("Explained variance ratio:", pca.explained_variance_ratio_)
print("Total variance explained:", pca.explained_variance_ratio_.sum())
```

# **Supervised Models**

### **Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Separate features and labels
X_train_ml = train_df.drop(columns=["activity", "activity_name"])
y_train_ml = train_df["activity"]

X_test_ml = test_df.drop(columns=["activity"])
y_test_ml = test_df["activity"]

# Initialize Logistic Regression
log_reg = LogisticRegression(max_iter=1000, n_jobs=-1)

# Train
log_reg.fit(X_train_ml, y_train_ml)

# Predict
y_pred_lr = log_reg.predict(X_test_ml)

# Evaluation
print("Logistic Regression Accuracy:", accuracy_score(y_test_ml, y_pred_lr))
print("\nClassification Report:\n", classification_report(y_test_ml, y_pred_lr))
```

```python
import seaborn as sns
import numpy as np

# Confusion matrix
cm = confusion_matrix(y_test_ml, y_pred_lr)

# Plot
plt.figure()
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=activity_map.values(),
            yticklabels=activity_map.values())

plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix - Logistic Regression")
plt.show()
```

## **SVM (Linear)**

```python
from sklearn.svm import SVC

# Linear SVM
svm_linear = SVC(kernel="linear")

# Train
svm_linear.fit(X_train_ml, y_train_ml)

# Predict
y_pred_svm_lin = svm_linear.predict(X_test_ml)

# Evaluate
print("Linear SVM Accuracy:", accuracy_score(y_test_ml, y_pred_svm_lin))
print("\nClassification Report:\n", classification_report(y_test_ml, y_pred_svm_lin))
```

```python

# Confusion matrix for Linear SVM
cm_svm_lin = confusion_matrix(y_test_ml, y_pred_svm_lin)

plt.figure()
sns.heatmap(cm_svm_lin, annot=True, fmt="d", cmap="Greens",
            xticklabels=activity_map.values(),
            yticklabels=activity_map.values())

plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix - Linear SVM")
plt.show()
```

# **Unsupervised Models**

## **Data Preprocessing For Unsupervised Models**

```python
# Data for clustering (no labels)
X_cluster = train_df.drop(columns=["activity", "activity_name"])

print("Clustering data shape:", X_cluster.shape)
```

## **KMeans Clustering**

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# K-Means
kmeans = KMeans(n_clusters=6, random_state=42)

kmeans_labels = kmeans.fit_predict(X_cluster)

# Evaluation
sil_score = silhouette_score(X_cluster, kmeans_labels)

print("K-Means Silhouette Score:", sil_score)
```

```python
# PCA for visualization
pca_cluster = PCA(n_components=2)
X_cluster_pca = pca_cluster.fit_transform(X_cluster)

# Create dataframe
cluster_df = pd.DataFrame(X_cluster_pca, columns=["PC1", "PC2"])
cluster_df["Cluster"] = kmeans_labels

# Plot clusters
plt.figure()
for c in sorted(cluster_df["Cluster"].unique()):
    subset = cluster_df[cluster_df["Cluster"] == c]
    plt.scatter(subset["PC1"], subset["PC2"], label=f"Cluster {c}", alpha=0.5)

plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("K-Means Clustering Visualization (PCA-reduced)")
plt.legend()
plt.show()
```

## **Agglomerative Clustering**

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score

# Agglomerative clustering
agg = AgglomerativeClustering(n_clusters=6, linkage="ward")

agg_labels = agg.fit_predict(X_cluster)

# Evaluation
agg_sil_score = silhouette_score(X_cluster, agg_labels)

print("Agglomerative Clustering Silhouette Score:", agg_sil_score)
```

```python
# PCA visualization for Agglomerative clustering
pca_cluster = PCA(n_components=2)
X_cluster_pca = pca_cluster.fit_transform(X_cluster)

agg_df = pd.DataFrame(X_cluster_pca, columns=["PC1", "PC2"])
agg_df["Cluster"] = agg_labels

plt.figure()
for c in sorted(agg_df["Cluster"].unique()):
    subset = agg_df[agg_df["Cluster"] == c]
    plt.scatter(subset["PC1"], subset["PC2"], label=f"Cluster {c}", alpha=0.5)

plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("Agglomerative Clustering Visualization (PCA-reduced)")
plt.legend()
plt.show()
```

## **DBSCAN- Additional**

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
import numpy as np

# DBSCAN clustering
dbscan = DBSCAN(eps=3.0, min_samples=10)

db_labels = dbscan.fit_predict(X_cluster)

# Count clusters and noise
unique_labels = set(db_labels)
n_clusters = len(unique_labels) - (1 if -1 in unique_labels else 0)
n_noise = np.sum(db_labels == -1)

print("Number of clusters found:", n_clusters)
print("Number of noise points:", n_noise)
```

```python

# PCA visualization for DBSCAN
pca_cluster = PCA(n_components=2)
X_cluster_pca = pca_cluster.fit_transform(X_cluster)

db_df = pd.DataFrame(X_cluster_pca, columns=["PC1", "PC2"])
db_df["Cluster"] = db_labels

plt.figure()
for label in sorted(db_df["Cluster"].unique()):
    subset = db_df[db_df["Cluster"] == label]
    if label == -1:
        plt.scatter(subset["PC1"], subset["PC2"], label="Noise", alpha=0.3)
    else:
        plt.scatter(subset["PC1"], subset["PC2"], label=f"Cluster {label}", alpha=0.6)

plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("DBSCAN Clustering Visualization (PCA-reduced)")
plt.legend()
plt.show()
```

# **Hyperparameter Tuning**

## **Supervised Cross Validation**

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

# Parameter grid
param_grid = {
    "C": [0.01, 0.1, 1, 10]
}

# Linear SVM
svm_linear = SVC(kernel="linear")

# Grid Search with Cross-Validation
grid_svm = GridSearchCV(
    svm_linear,
    param_grid,
    cv=5,
    scoring="accuracy",
    n_jobs=-1
)

# Fit
grid_svm.fit(X_train_ml, y_train_ml)

# Results
print("Best C:", grid_svm.best_params_)
print("Best CV Accuracy:", grid_svm.best_score_)
```

```python
# Train final model with best C
best_svm = SVC(kernel="linear", C=1)
best_svm.fit(X_train_ml, y_train_ml)

# Test performance
y_pred_best = best_svm.predict(X_test_ml)

print("Test Accuracy (Best Linear SVM):", accuracy_score(y_test_ml, y_pred_best))
print("\nClassification Report:\n", classification_report(y_test_ml, y_pred_best))
```

```python
# CV vs Test accuracy visualization
cv_accuracy = grid_svm.best_score_
test_accuracy = accuracy_score(y_test_ml, y_pred_best)

plt.figure()
plt.plot(["Cross-Validation", "Test"], [cv_accuracy, test_accuracy], marker="o")
plt.ylabel("Accuracy")
plt.title("Linear SVM: Cross-Validation vs Test Accuracy")
plt.ylim(0.9, 1.0)
plt.show()

print("CV Accuracy :", cv_accuracy)
print("Test Accuracy:", test_accuracy)
```

```python
# Accuracy values
default_svm_accuracy = accuracy_score(y_test_ml, y_pred_svm_lin)
tuned_svm_accuracy = accuracy_score(y_test_ml, y_pred_best)

# Comparison table
svm_comparison = pd.DataFrame({
    "Model": [
        "Linear SVM (Default)",
        "Linear SVM (Tuned, C=1)"
    ],
    "Accuracy": [
        default_svm_accuracy,
        tuned_svm_accuracy
    ]
})

svm_comparison
```

## **Unsupervised Model Fine - Tuning**

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

k_values = range(2, 11)
silhouette_scores_k = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X_cluster)
    score = silhouette_score(X_cluster, labels)
    silhouette_scores_k.append(score)

# Put results in DataFrame
kmeans_tuning = pd.DataFrame({
    "k": k_values,
    "Silhouette Score": silhouette_scores_k
})

kmeans_tuning
```

```python
# Default vs tuned K-Means comparison
kmeans_comparison = pd.DataFrame({
    "Model": [
        "K-Means (k = 6, activity-based)",
        "K-Means (k = 2, silhouette-optimal)"
    ],
    "k": [6, 2],
    "Silhouette Score": [
        kmeans_tuning[kmeans_tuning["k"] == 6]["Silhouette Score"].values[0],
        kmeans_tuning[kmeans_tuning["k"] == 2]["Silhouette Score"].values[0]
    ]
})

kmeans_comparison
```

```python
plt.figure()
plt.plot(kmeans_tuning["k"], kmeans_tuning["Silhouette Score"], marker="o")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("K-Means Hyperparameter Tuning using Silhouette Score")
plt.show()
```
